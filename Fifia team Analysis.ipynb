```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("fifa_data.csv",delimiter=",")
```

```python
barce = data.loc[data["Club"]== "FC Barcelona"]
# barce["Potential"]
# np.sort(barce)
```

```python
plt.plot(barce["Potential"],barce["Release Clause"],"r.--",label="Player")
plt.title("Fifa Players Release Clause Vs Potential")
plt.xlabel("Player Potential")
plt.ylabel("Player Release Clause")
plt.yticks(barce["Release Clause"].to_list()[::2])
plt.legend()
plt.savefig("Fifa Players Release Clause Vs Potential",format="png",dpi=500)
plt.show()
```

```python
# bins = [100,200,300,400,500,600,700,800,900]
plt.hist(data["Overall"],color= "#a9f76d",label = "Player Skill Level")
plt.title("Players Distribution w.r.t Skill Level in Fifa")
plt.xlabel("Number Of Players")
plt.ylabel("Skill Level")
plt.legend()
plt.savefig("Players Distribution w.r.t Skill Level in Fifa",format="png",dpi=500)
plt.show()
```

```python
left = data.loc[data["Preferred Foot"] == "Left"].count()[0]
right = data.loc[data["Preferred Foot"] == "Right"].count()[0]
```

```python
plt.pie([left,right],labels = ["Left","Right"],colors=["#a9f76d","#6df7d5"],autopct="%.2f%%")
plt.title("Left and Right Foot Player Analysis")
plt.savefig("Left and Right Foot Player Analysis",format="png",dpi=500)
plt.show()
```

```python
weight = pd.Series([int(x.strip("lbs")) if type(x) == str else x for x in data.Weight])
# weight((weight)
```

```python
light = weight.loc[weight<125].count()
light_medium = weight.loc[(weight>=125) & (weight< 150)].count()
medium = weight.loc[(weight>=150) & (weight <175)].count()
medium_heavy = weight.loc[(weight>=175) & (weight<200)].count()
heavy = weight.loc[weight >= 200].count()

heavy
```

```python
weights =[light,light_medium,medium,medium_heavy,heavy]
labels = ["Under 125","125-149","150-174","175-199","Above 200"]
explode = [0.5,0,0,0,0.5]
plt.style.use('ggplot')
plt.pie(weights,wedgeprops={'linewidth':0.2,'edgecolor': 'white'}, labels = labels, explode = explode, autopct="%.2f%%" )
plt.title("Weight Distribution of FIFA")
plt.savefig("Weight Distribution of FIFA",format="png",dpi=500)
plt.show()
```

```python
clubs = data["Club"]
clubs = list(clubs.unique())
teams_potential = []
for club in clubs:
    teams_potential.append(data.loc[data["Club"] == club ]["Overall"])


barca = data.loc[data["Club"] == "FC Barcelona" ]["Overall"]
madrid = data.loc[data["Club"] == "Real Madrid" ]["Overall"]
manchester_city = data.loc[data["Club"] == "Manchester City" ]["Overall"]


plt.boxplot([barca,madrid,manchester_city],labels = ["FC Barcelona","Real Madrid","Manchester City"])
plt.title("Club Comparision")
plt.xlabel("Clubs")
plt.ylabel("Over all Potential")
plt.savefig("Club Comparision",format="png",dpi=500)
plt.show()
```
